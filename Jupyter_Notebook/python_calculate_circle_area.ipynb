# Drawing Triangle with circle

```python
import math
```

```python
def drawTriangle(r, angleStart, angleEnd):
    p1 = [0,0]
    rs = math.radians(angleStart)
    re = math.radians(angleEnd)
    p2 = [math.cos(rs), math.sin(rs)]
    p3 = [math.cos(re), math.sin(re)]
    points = [p1, p2, p3, p1]
    x, y = zip(*points)
    plt.plot(x, y)

def drawing(split):
    start = 0
    rotate = (360 / split)
    for i in range(split):
        drawTriangle(1, start, start + rotate)
        start += rotate
    plt.axis([-2, 2, -2, 2])
    plt.title('Split to :{}'.format(split))
    plt.show()

steps = [4, 8, 16, 32]
for step in steps:
    drawing(step)
```

# Calculate Area

```python
a = 15
r = math.radians(a)
adj = math.cos(r)
opp = math.sin(r)

# triangle A
p1 = [0,0]
p2 = [adj, 0]
p3 = [adj, opp]
points = [p1, p2, p3, p1]
x, y = zip(*points)
plt.plot(x, y, label='A')

# triangle B
p1 = [0,0]
p2 = [adj, 0]
p3 = [adj, -opp]
points = [p1, p2, p3, p1]
x, y = zip(*points)

plt.plot(x, y, label='B')
plt.legend(loc="lower right")
plt.axis([0, 1, -0.5, 0.5])
```

#  Do Slice

```python
def getArea(r, angle):
    pr = math.radians(angle/2)
    opp = math.sin(pr) * r
    adj = math.cos(pr) * r
    return opp * adj
```

```python
for step in steps:
    angle = 360 / step
    area = getArea(1, angle) * step
    print('slice : {} area is : {}'.format(step, area))
```

```python
steps = [128, 1024, 32768, 65536, 1048576]
for step in steps:
    angle = 360 / step
    area = getArea(1, angle) * step
    print('slice : {} area is : {}'.format(step, area))
```
